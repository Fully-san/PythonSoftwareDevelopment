## Créer une calculatrice

Ecrivez un programme réalisant une calculatrice pouvant faire les opérations suivantes :
-	Addition
-	Soustraction
-	Multiplication
-	Division
-	Reste de la division euclidienne

Pour ce faire vous ferez les opérations suivantes :
-	Demander à l’utilisateur d’entrer un nombre (un flottant)
-	Demander à l’utilisateur d’entrer un opérateur (un caractère)
-	Demander à l’utilisateur d’entrer un nombre (un flottant)
-	Effectuer le calcul


<details>
<summary>Solution</summary>
<p>
num1 = eval(input("Entrer un nombre : "))
operator = input("Entrer une operation (+ - * / %) : ")
num2 = eval(input("Entrer un nombre : "))

if operator == '+':
    result = num1 + num2
elif operator == '-':
    result = num1 - num2
elif operator == '*':
    result = num1 * num2
elif operator == '/':
    result = num1 / num2
elif operator == '%':
    result = num1 % num2
else:
    print("Il y a une erreur de saisie.")

print("Le resultat de l\'operation est :", result)
