#### La fonction qui va définir le nombre d'allumette de la partie

```python
def defineMatchstickNumber():
    matchstickNumber = 0

    while matchstickNumber % 2 == 0:
        matchstickNumber = int(input("Combien d'allumettes voulez vous : "))

    return matchstickNumber

matchstickNumber = defineMatchstickNumber()
```

*texte en italique*#### La fonction qui détermine qui commence

```python
COMPUTER = 0
HUMAN = 1

def whoStart():
    player = -1

    while player != COMPUTER and player != HUMAN:
        player = int(input(f"Qui commence ({COMPUTER} pour l'ordi, {HUMAN} pour vous) : "))

    return player

player = whoStart()
```

```python
import random

def removeMatchstickNumber(matchstickNumber):
    choice = 0

    if matchstickNumber > 3:
        choice = random.randint(1, 3)
    else:
        choice = matchstickNumber

    print(f"Il y a {matchstickNumber} allumettes, j'en enlève {choice}.")

    return matchstickNumber - choice

matchstickNumber = removeMatchstickNumber(matchstickNumber)
```

#### La boucle de jeu

```python
def game():
    matchstickNumber = defineMatchstickNumber()
    player = whoStart()

    while matchstickNumber > 0:
        matchstickNumber = removeMatchstickNumber(matchstickNumber, player)
        player = not player

game()
```

#### La fonction de victoire

```python
def displayResult(player):
    if player == COMPUTER:
        print("Vous avez gagné !")
    else:
        print("Vous avez perdu...")
```
